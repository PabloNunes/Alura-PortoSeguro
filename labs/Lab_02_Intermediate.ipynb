## Exercise 1: Creating Functions

Learn how to create and use functions in Python.

```python
def calculate_average(numbers):
    """Calculate the average of a list of numbers."""
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

# Test the function
grades = [85, 92, 78, 90, 88]
avg = calculate_average(grades)
print(f"Average grade: {avg:.2f}")
```

## Exercise 2: Working with Dictionaries

Practice using Python dictionaries for data management.

```python
# Create a dictionary with student information
student = {
    "name": "João Silva",
    "age": 28,
    "course": "Data Science",
    "grades": [85, 92, 78, 90, 88]
}

# Access and display information
print(f"Student: {student['name']}")
print(f"Course: {student['course']}")
print(f"Average grade: {calculate_average(student['grades']):.2f}")
```

## Exercise 3: Data Analysis with Lists

Perform basic statistical analysis on data.

```python
# Sample data: monthly sales
sales = [12000, 15000, 13500, 18000, 16500, 14000, 19000, 17500, 20000, 18500, 21000, 22000]

# Calculate statistics
total_sales = sum(sales)
average_sales = total_sales / len(sales)
max_sales = max(sales)
min_sales = min(sales)

print(f"Total Sales: R$ {total_sales:,.2f}")
print(f"Average Monthly Sales: R$ {average_sales:,.2f}")
print(f"Best Month: R$ {max_sales:,.2f}")
print(f"Worst Month: R$ {min_sales:,.2f}")
```

## Exercise 4: List Comprehension

Practice using list comprehension for efficient data manipulation.

```python
# Create a list of squared numbers
numbers = range(1, 11)
squares = [n**2 for n in numbers]
print(f"Squares: {squares}")

# Filter even numbers
even_squares = [n**2 for n in numbers if n % 2 == 0]
print(f"Even squares: {even_squares}")
```
