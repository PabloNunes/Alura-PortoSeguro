## Exercise 1: Exception Handling

Learn how to handle errors gracefully in Python.

```python
def safe_divide(a, b):
    """Safely divide two numbers with error handling."""
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
        return None
    except TypeError:
        print("Error: Invalid input types!")
        return None

# Test the function
print(f"10 / 2 = {safe_divide(10, 2)}")
print(f"10 / 0 = {safe_divide(10, 0)}")
print(f"10 / 'a' = {safe_divide(10, 'a')}")
```

## Exercise 2: Object-Oriented Programming

Create and work with classes in Python.

```python
class Customer:
    """Represents a customer in the insurance system."""
    
    def __init__(self, name, age, policy_number):
        self.name = name
        self.age = age
        self.policy_number = policy_number
        self.claims = []
    
    def add_claim(self, amount, description):
        """Add a new claim for this customer."""
        claim = {
            "amount": amount,
            "description": description
        }
        self.claims.append(claim)
    
    def get_total_claims(self):
        """Calculate total claim amount."""
        return sum(claim["amount"] for claim in self.claims)
    
    def display_info(self):
        """Display customer information."""
        print(f"Customer: {self.name}")
        print(f"Age: {self.age}")
        print(f"Policy Number: {self.policy_number}")
        print(f"Total Claims: R$ {self.get_total_claims():,.2f}")
        print(f"Number of Claims: {len(self.claims)}")

# Create a customer instance
customer = Customer("Maria Santos", 35, "PS-2024-001")
customer.add_claim(1500.00, "Car accident repair")
customer.add_claim(800.00, "Windshield replacement")
customer.display_info()
```

## Exercise 3: Working with Data Structures

Advanced data manipulation and analysis.

```python
# Sample insurance data
policies = [
    {"id": "PS-001", "premium": 1200, "coverage": 50000, "type": "Auto"},
    {"id": "PS-002", "premium": 2500, "coverage": 100000, "type": "Home"},
    {"id": "PS-003", "premium": 1800, "coverage": 75000, "type": "Auto"},
    {"id": "PS-004", "premium": 3000, "coverage": 150000, "type": "Home"},
    {"id": "PS-005", "premium": 1500, "coverage": 60000, "type": "Auto"},
]

# Calculate statistics
total_premium = sum(policy["premium"] for policy in policies)
total_coverage = sum(policy["coverage"] for policy in policies)
avg_premium = total_premium / len(policies)

# Group by type
auto_policies = [p for p in policies if p["type"] == "Auto"]
home_policies = [p for p in policies if p["type"] == "Home"]

print(f"Total Policies: {len(policies)}")
print(f"Total Premium: R$ {total_premium:,.2f}")
print(f"Total Coverage: R$ {total_coverage:,.2f}")
print(f"Average Premium: R$ {avg_premium:,.2f}")
print(f"\nAuto Policies: {len(auto_policies)}")
print(f"Home Policies: {len(home_policies)}")
```

## Exercise 4: Lambda Functions and Higher-Order Functions

Use functional programming concepts in Python.

```python
# Sample risk scores
customers_risk = [
    {"name": "João", "risk_score": 75},
    {"name": "Maria", "risk_score": 45},
    {"name": "Pedro", "risk_score": 90},
    {"name": "Ana", "risk_score": 30},
    {"name": "Carlos", "risk_score": 60},
]

# Sort by risk score
sorted_by_risk = sorted(customers_risk, key=lambda x: x["risk_score"], reverse=True)
print("Customers sorted by risk (highest first):")
for customer in sorted_by_risk:
    print(f"{customer['name']}: {customer['risk_score']}")

# Filter high-risk customers (score > 70)
high_risk = list(filter(lambda x: x["risk_score"] > 70, customers_risk))
print(f"\nHigh-risk customers: {[c['name'] for c in high_risk]}")

# Apply discount based on risk
apply_discount = lambda score: 1.0 if score < 50 else (0.9 if score < 75 else 0.8)
discounts = list(map(lambda c: {"name": c["name"], "discount": apply_discount(c["risk_score"])}, customers_risk))
print("\nDiscount multipliers:")
for item in discounts:
    print(f"{item['name']}: {item['discount']:.0%}")
```
